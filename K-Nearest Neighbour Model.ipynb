```python
import numpy as np
import pandas as pd

column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/housing.csv', delimiter=r"\s+", names=column_names)

dataset.head()
```

**Data Preparation and Split**

```python
# Data is from the Linear Regression Model is its clean already.

from sklearn.model_selection import train_test_split
X = np.array(dataset.iloc[:, 0:13])
y = np.array(dataset['MEDV'])

# training for 70-30 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

**Import KNN Model : Regression Type from Sklearn**

```python
from sklearn.neighbors import KNeighborsRegressor

# load KNN model
kn_model = KNeighborsRegressor(3)  # 3 is the K value
kn_model.fit(X_train, y_train)
kn_predict = kn_model.predict(X_test)
```

**HyperParameter Tuning (K)**

```python
from sklearn.metrics import mean_squared_error

for i in range(1, 50):     # I decide I wanna try for k =1 to k = 50, its a small dataset
  model = KNeighborsRegressor(i)    # so using a loop is not that performance heavy
  model.fit(X_train, y_train)
  model_predict = model.predict(X_test)
  # Mean Squared Error
  mse = mean_squared_error(y_test, model_predict, squared=False)

  print(f'{mse} is the Mean-Sqaured Error for k = {i}')
```

**k = 4 is the least MSE for this Model : 7.004**

**At k = 3 the MSE is 7.014927171138291**

Therefore, K-Nearest Neighbor Model performs ~27% Worse than Linear Regression at k = 3 and and ~26% at k = 4 (Best Case) **for this Boston Housing Dataset**
