```python
# Show percentage on countplot/barplot
def barPerc_without_hue(ax, feature):
    total = len(feature)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width()
        y = p.get_y() + p.get_height()
        ax.annotate(percentage, (x, y), size = 12)
```

```python
X = df1[['Color', 'Model', 'EOL Flag', 'Price', 'Rating', 'Total Sales Per Product']]
y = df1['Category']
```

```python
# Encode the categorical features
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
encoder = LabelEncoder()
for col in X.columns:
    if X[col].dtype == 'object':
        X[col] = encoder.fit_transform(X[col])

# Convert the Pandas DataFrame to a NumPy array
X = X.to_numpy()

# Reshape the data to a 2D array
X = X.reshape(-1, 1)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)


# Create a recommendation model
clf = KNeighborsClassifier(n_neighbors=5)

# Train the model
clf.fit(X_train, y_train)

# Evaluate the model
y_pred = clf.predict(X_test)
accuracy = (y_pred == y_test).mean()
print('Accuracy:', accuracy)

# Use the model to make recommendations
def recommend(product_features, products):
    # Calculate the distances between the product features and the features of the products in the products DataFrame
    distances = clf.kneighbors(product_features, return_distance=True)[0]

    # Get the indices of the products with the smallest distances
    recommendation_indices = distances.argsort()[:5]

    # Get the recommended products
    recommended_products = products.iloc[recommendation_indices]

    return recommended_products

# Get the features of the product to recommend
product_features = X.iloc[0]

# Get the recommended products
recommended_products = recommend(product_features, df1)

# Print the recommended products
print(recommended_products)
```
