**Libraries and Dataset Import**

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import sklearn as sk

column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/housing.csv', delimiter=r"\s+", names=column_names)
```

**Splitting Data for Training and Testing**

```python
from sklearn.model_selection import train_test_split
X = np.array(dataset.iloc[:, 0:13]) # give me all rows and first 12 columns
Y = np.array(dataset['MEDV']) # MEDV is the output variable

# setting test/validation data to be 30% of original data
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=5)

# train_test_split took 4 parameters, input vars , output, test size (1 is 100%)
# and random_state??? its a random seed, basically in simple words, it ensures the data is always split in the same way no matter how many
# times you run to train or test the model.

```

**Importing the Linear Regression Model of Sklearn**

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

# load our first model
lr_model = LinearRegression()

lr_model.fit(x_train, y_train)

predict_lr = lr_model.predict(x_test)
```

**Calculating Mean Squared Error for Model Evalution**

```python
mse_lr = sk.metrics.mean_squared_error(y_test, predict_lr, squared=False)
print(f'The Error for this Linear Regression Model is {mse_lr}')
```
