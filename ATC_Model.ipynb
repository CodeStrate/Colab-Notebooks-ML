```python
import swifter

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
import re, nltk, spacy, string
model = spacy.load("en_core_web_sm")
import en_core_web_sm
nlp = en_core_web_sm.load()

# corpus wordnet
nltk.download("wordnet")
# corpus punkt
nltk.download("punkt")
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
from nltk.corpus import stopwords

from nltk import pos_tag, word_tokenize

from nltk.stem import WordNetLemmatizer

from gensim.corpora.dictionary import Dictionary
from gensim.models.nmf import Nmf
from gensim.models.coherencemodel import CoherenceModel
from operator import itemgetter

import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np

from textblob import TextBlob
```

```python
df = pd.read_json('/content/drive/MyDrive/Colab Notebooks/Datasets/ticket_cleaned.json')
dfc = df.drop(columns=['issue'])

dfc.rename(columns={'complaint_what_happened' : 'complaints'}, inplace=True)


```

```python
df.sample(20)
```

```python
def cleaner(data):
  data = data.lower()
  data = re.sub(r'\[|\]',"",data)    # Remove text in square brackets
  data = re.sub(r'[^\w\s]',"",data)  # Remove punctuation
  data = re.sub(r'\w*\d\w*',"",data) # Remove words containing numbers
  return " ".join(data.split())  # Remove unwanted empty spaces

dfc['complaints'] = dfc['complaints'].swifter.apply(cleaner)
```

```python
dfc['complaints'].iloc[1]
```

```python
def checkFrequency(text):
  return len(str(text).split(' ')) # sentence to list

dfc['wordFrequency'] = dfc['complaints'].swifter.apply(checkFrequency)

dfc['wordFrequency'].describe()
```

```python
dfc.head(5)
```

```python
def lemmatize_text(text):
  lemmatized =[]
  doc = nlp(text)
  for word in doc:
    lemmatized.append(word.lemma_)
  return " ".join(lemmatized)

def get_nouns(text):
  blob = TextBlob(text)
  return ' '.join([ word for (word,tag) in blob.tags if tag == "NN"])


dfc['lemmatized complaints'] = dfc['complaints'].swifter.apply(lemmatize_text)
```

```python
dfc['lemmatized complaints'] = dfc['lemmatized complaints'].swifter.apply(get_nouns)
```

```python
df_clean = dfc[['complaints', 'lemmatized complaints']].copy()

df_clean.head()
```

```python
def extract_POS_tags(data):
  nouns = [token for token, pos in pos_tag(word_tokenize(data)) if pos.startswith('N')]
  return ' '.join(nouns)

df_clean['complaints_POS_removed'] = dfc['lemmatized complaints'].swifter.apply(extract_POS_tags)
```

```python
df_clean.head()
```

```python
df_clean['complaintsClean'] = df_clean['complaints_POS_removed'].str.replace('xxxx','')

def remove_allXX(text):
  return re.sub('[x]{2,}',"",text)

df_clean['complaintsClean'] = df_clean['complaintsClean'].swifter.apply(remove_allXX)
```

```python
df_clean.sample(10)
```

**FEATURE EXTRACTION : USING NON-NEGATIVE MATRIX FACTORIZATION (NMF)**

```python
tfidf = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english') # 0.95 = 95% , ignore words repeated in 95% of complaints,

# 2 for min_df is the lower bound, ignore words infrequent, appearing in less than 2 complaints
```

```python
# create a doc term matrix : keeping track of words appearing in complaints, it wont have words that have a score of 0 , words that
# only appear in one complaint . that way we can find the most dominant terms in complaints and categorize based on that

dtm = tfidf.fit_transform(df_clean['complaintsClean'])
```

Topic Modelling : extracting words from complaints using nmf

```python
from sklearn.decomposition import NMF

# Use gensim's nmf model here

texts = df_clean['complaintsClean']

dataset = [d.split() for d in texts]

# Create a Dictionary to keep track of word occurance count

dictionary = Dictionary(dataset)

# filter out outliers

dictionary.filter_extremes(
    no_below=3,
    no_above=0.85,
    keep_n=5000
)


# create bag of words (token : count)

texts_corpus = [dictionary.doc2bow(text) for text in dataset]

# random issue topic list
topic_nums = list(np.arange(5, 10, 1)) # 5,6,7,8,9

coherence_scores = []

for num in topic_nums:
    nmf = Nmf(
        corpus=texts_corpus,
        num_topics=num,
        id2word=dictionary,
        chunksize=2000,
        passes=5,
        kappa=.1,
        minimum_probability=0.01,
        w_max_iter=300,
        w_stop_condition=0.0001,
        h_max_iter=100,
        h_stop_condition=0.001,
        eval_every=10,
        normalize=True,
        random_state=42
    )

    # run coherence model here
    cm = CoherenceModel(
        model=nmf,
        texts=texts,
        dictionary=dictionary,
        coherence='c_v'
    )

    coherence_scores.append(round(cm.get_coherence(), 5))

scores = list(zip(topic_nums, coherence_scores)) # zip maps both lists into one score list
best_num_issue_topics = sorted(scores, key=itemgetter(1), reverse=True)[0][0]

print(best_num_issue_topics)
```

**Hence 5 topics are the best to classify this data**

```python
# load sklearn's nmf model for trial and error approach

num_topics = 5 # say 5

# here random_state = 40 say
nmf_model = NMF(n_components=num_topics, random_state=40)
```

```python
nmf_model.fit(dtm)
len(tfidf.get_feature_names_out())
```

```python
for index,topic in enumerate(nmf_model.components_):
    print(f'THE TOP 15 WORDS FOR TOPIC #{index}')
    print([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-15:]])
    print('\n')
```

```python
issue_results = nmf_model.transform(dtm)
issue_results[0].round(2)
issue_results[0].argmax()
issue_results.argmax(axis=1)
```

```python
# assign this to a column

df_clean['Issue Index'] = issue_results.argmax(axis=1)
```

```python
df_clean = df_clean.groupby('Issue Index').head(8)
df_clean.sort_values('Issue Index')
```

```python
Issue_Dictionary = {
    0 : "Bank account and/or its services Issue",
    1 : "Credit card related Issue",
    2 : "Other Issues",
    3 : "Fraud or theft disputes",
    4 : "Mortgage and Loan Issue"
}

# map issue dict to column issues

df_clean['Issue_Category'] = df_clean['Issue Index'].map(Issue_Dictionary)

df_clean
```

```python
training_data = df_clean[['complaintsClean', 'Issue Index']]

training_data
```

**split data and create models**

```python
X = training_data['complaintsClean']
y = training_data['Issue Index']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.75, stratify = y, random_state = 100)
```

```python
TF_IDF_vectorizer = TfidfVectorizer(stop_words = "english", ngram_range = (1, 3), max_df = 0.95, min_df = 2, max_features = 10000)
X_train = TF_IDF_vectorizer.fit_transform(X_train).toarray()
X_test = TF_IDF_vectorizer.transform(X_test).toarray()
print(len(TF_IDF_vectorizer.get_feature_names_out()))
```

```python
y_train.value_counts()/len(y_train)
```

```python
model_1 = LogisticRegression(class_weight = "balanced", random_state = 100, max_iter = 500, n_jobs = -1)
model_1.fit(X_train, y_train)
```

```python
def model_evaluation(model) :
  y_train_pred = model.predict(X_train)
  y_test_pred = model.predict(X_test)
  print("Accuracy score on train data :", accuracy_score(y_train, y_train_pred))
  print("Accuracy score on test data :", accuracy_score(y_test, y_test_pred), end = "\n\n")
  print("Confusion matrix for test data :", confusion_matrix(y_test, y_test_pred), end = "\n\n", sep = "\n")
  print("Classification report on test data  :", classification_report(y_test, y_test_pred), sep = "\n")
```

**Logistic Regression Model**

```python
model_evaluation(model_1)
```

```python
model_2 = MultinomialNB()
model_2.fit(X_train, y_train)
```

**Multinomial Naive Bayes Model**

```python
model_evaluation(model_2)
```
