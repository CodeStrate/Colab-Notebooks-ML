```python
# 1. Normalizing the Data
# Write a code which works exactly like normalize(data).

import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(data = iris.data, columns=iris.feature_names)
cols = list(df.columns)
cols
```

```python
def minmaxer(df, cols):
  for col in cols:
    minimum = df[col].min()
    maximum = df[col].max()
    df[col] = (df[col] - minimum) / (maximum - minimum)
  return df

df = minmaxer(df, cols)

df.head(5)
```

```python
# 2. Cost Function

def cost_function(X, y, theta):
    m = len(y)
    h_theta_Xi = np.dot(X, theta)
    squared_errors = (h_theta_Xi - y) ** 2
    cost = (1 / (2 * m)) * np.sum(squared_errors)
    return cost

theta = np.array([1.0, 2.0])
X = np.array([[1.0, 3.0], [2.0, 4.0]])
y = np.array([5.0, 6.0])
X = X.T

cost = cost_function(X, y, theta)
print("Cost:", cost)



```

```python
# 3. Mapping Variables

l1 = ["Reetesh", "Shruti", "Kaustubh", "Vikas", "Mahima", "Akshay"]
l2 = ['No', 'Maybe', 'Yes', 'yes', 'maybe', 'yes']

l2_opt = [x.lower() for x in l2]

zipped = list(zip(l1, l2_opt))

resp_dict = {'yes' : 1.0, 'no' : 0.0, 'maybe' : 0.5}

df = pd.DataFrame(zipped, columns=['Name', 'Response'])

df['Response'] = df['Response'].map(resp_dict)

print(df)
```

```python
# 8. Evaluation Metrics

def max_stat(df):
  tp = df[0][0] # 50
  tn = df[1][0] # 80
  fp = df[1][1] # 40
  fn = df[0][1] # 30

  accuracy = (tp + tn) / np.sum(df)
  senstivity = tp / (tp + fn)
  specificity = tn / (tn + fp)


  return [accuracy, senstivity, specificity]


df = ((50, 30), (80, 40))
res = max_stat(df)
print(res)

print(f'Max stat for df is {max(res)}')



```

```python
# F1 Score

def f1_score(df):
  tp = df[0][0]
  tn = df[1][0]
  fp = df[1][1]
  fn = df[0][1]

  precision = tp / (tp + fp)
  recall = tp / (tp + fn)

  f1 = (2 *recall*precision)/(recall + precision)

  return f1

df = ((150, 50),(400, 100))
print(f1_score(df))
```
