```python
import numpy as np
import pandas as pd
```

```python
# Weights Calculation

# Function to calculate the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Function to perform batch gradient descent
def gradient_descent(X, y, weights, learning_rate):
    m = len(y)
    predictions = sigmoid(np.dot(X, weights))
    error = predictions - y
    gradient = np.dot(X.T, error) / m
    weights -= learning_rate * gradient
    return weights

# Input data
X = np.array([[10, -0.5, 1], [55, 0.3, 0]])
y = np.array([0, 1])

# Initial weights for the 20th iteration
weights_20th = np.array([0.04, -1, 0.4])

# Learning rate
learning_rate = 0.2

# Calculate weights for the 21st iteration
weights_21st = gradient_descent(X, y, weights_20th, learning_rate)

# Print the weights of the 21st iteration
print("Weights in the 21st iteration:", weights_21st)
```

```python
# Predicting Class Label

# Function to calculate the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Given weights
weights = np.array([-0.025, -0.957, 0.322])

# Data point features
data_point = np.array([40, 0.6, 1])

# Calculate the predicted probability (σ)
predicted_probability = sigmoid(np.dot(data_point, weights))

# Threshold
threshold = 0.6

# Classify based on the threshold
predicted_class = 1 if predicted_probability >= threshold else 0

# Print the results
print("Predicted Probability (σ):", predicted_probability)
print("Predicted Class:", predicted_class)
```

```python
# Bayes Theorem

# Given information
prob_bag_A = 0.5  # Probability of choosing Bag A
prob_bag_B = 0.5  # Probability of choosing Bag B

prob_green_given_A = 4 / 7  # Probability of drawing a green ball given Bag A
prob_green_given_B = 6 / 10  # Probability of drawing a green ball given Bag B

# Bayes' Theorem
prob_A_given_green = (prob_bag_A * prob_green_given_A) / ((prob_bag_A * prob_green_given_A) + (prob_bag_B * prob_green_given_B))

# Print the result
print("Probability of choosing Bag A given a green ball was drawn:", round(prob_A_given_green, 2))
```

```python
# Hopkins Statistic

def hopkins(X):
    """
    Hopkins statistic function for checking the clustering tendency of a dataset.

    Parameters:
        X (pandas.DataFrame): The dataset to be tested.

    Returns:
        float: The Hopkins statistic value.
    """
    # Calculate the average nearest neighbor distance within the dataset
    d1 = np.mean([np.min(np.linalg.norm(X - x, axis=1)) for x in X])

    # Generate a uniform random dataset with the same size as the original dataset
    X_rand = np.random.rand(X.shape[0], X.shape[1])

    # Calculate the average nearest neighbor distance within the random dataset
    d2 = np.mean([np.min(np.linalg.norm(X_rand - x, axis=1)) for x in X_rand])

    # Check for division by zero
    if d1 == 0 or (d1 + d2) == 0:
        return np.nan

    # Calculate the Hopkins statistic
    H = 1 - (d1 / (d1 + d2))

    return H

# Example usage with Iris dataset
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data

H = hopkins(X)
print(H)
```
